# 主題:利用人臉辨識判斷我們與哪個明星最像

## 組員: 吳祐炘 R07941094、陳韋廷 R07941058

## 介紹:
### 現今許多人對於自身的外貌很有自信，常與人較勁長相得分高低，更是藉由明星來比喻自己，因此我們希望開發一個小程式，使用卷積神經網路CNN識別任一臉部直接與特定明星做連結。


## 實作方法:

### 1.進入kaggle裡將 CELEBA中200000筆資料下載下來
### 2.將圖片檔匯入，進行資料前處理。
### 3.建立CNN圖像辨識神經網絡，找到最高準確性並預測
### 4.input為我們個人大頭照，output為與我們相似的明星臉
